#### 1. Addition, Multiplication and Dot product of "Two Vectors"

```python
import numpy as np
vector1 = np.array([5,7,3,1])
print("One vector is : ",vector1)
vector2 = np.array([2,4,6,9])
print("Another vector is : ",vector2)
```

```python
print("Addition of 2 vectors : ", np.add(vector1,vector2))
```

```python
print("Subtraction of 2 vectors : ", np.subtract(vector1,vector2))
```

```python
print("Multiplication of 2 vectors : ", np.multiply(vector1,vector2))
```

```python
print("Dot product of 2 vectors : ", np.dot(vector1,vector2))
```

#### 2. Addition, Multiplication and Dot product of "Two Matrices"

```python
import numpy as np 
# A function to create a matrix
def fun():
    row = int(input("Enter number of rows"))
    column = int(input("Enter number of rows"))

    print("Enter the {} values in a single line 'separated by space': ".format(row*column)) 
    entries = list(map(int, input().split())) 
    matrix = np.array(entries).reshape(row, column) 
    return matrix 
```

```python
matrix1 = fun()
matrix1
```

```python
matrix2 = fun()
matrix2
```

```python
print("Addition of 2 matrix")
print(np.add(matrix1, matrix2))
```

```python
print("Subtraction of 2 matrix")
print(np.subtract(matrix1, matrix2))
```

```python
print("Multiplication of 2 matrix")
print(np.matmul(matrix1, matrix2))
```
